# import packages

```python
import pandas as pd
# import numpy as np
# from datetime import datetime
# import pytz
import model
import preprocess as prep
```

# load data set (training & test)

## load training set

```python
train = pd.read_csv("../../input/train.csv")
train["count"] = 1
train["train_test"] = "train"
train
```

 ## load test set

```python
test = pd.read_csv("../../input/test.csv")
test["count"] = 1
test["train_test"] = "test"
test
```

# data preprocessing

```python
print("train:\n",train.dtypes,"\n")
```

```python
print("test:\n",test.dtypes,"\n")
```

```python
## features to use for modeling
feature_list = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket_len', 'Fare', 'Cabin_len', 'Embarked', 'title', 'last_name', 'relatives', 'alone']
categorical = ['Pclass', 'Sex', 'alone', 'Embarked', 'title', 'last_name']
continuous = ['Age', 'SibSp', 'Parch', 'Ticket_len', 'Fare', 'Cabin_len', 'relatives']
target = "Survived"
PassengerId="PassengerId"
```

## concat training and test dataset for preprocessing.

```python
tr_te = pd.concat([train, test], sort=False)
tr_te
```

```python
feature_list_training = categorical + continuous
feature_list_train = feature_list + [target]
feature_list_test = feature_list + [PassengerId]

df = prep.preprocess(tr_te)

train = df.loc[df["train_test"]=="train",feature_list_train]
test = df.loc[df["train_test"]=="test",feature_list_test]

X_train = train[feature_list]
y_train = train[target]
X_test = test[feature_list]

display(train)
display(test)
```

## pipeline for preprocessing

```python
preprocessor = model.pipeline_preprocess(continuous, categorical)
preprocessor
```

# model1 : XGBOOST

## Bayesian Optimazation to find best hyperparameter for xgboost

```python
pbounds_xgb = {
    'learning_rate': (0.01, 1.0),
    'n_estimators': (10, 500),
    'max_depth': (3,7),
    'gamma': (0, 5)}

best_param_xgb = model.bayesian_opt_xgboost(X_train, y_train, pbounds_xgb, preprocessor)
```

## fit and predict using xgboost with best hyperparameter.

```python
res_tb_xgb = test[["PassengerId"]]
res_tb_xgb["Survived"] = model.xgb_fit_and_predict(preprocessor, best_param_xgb, X_train, y_train, X_test)
res_tb_xgb
```

```python
current_ts = model.current_timestamp()
res_tb_xgb.to_csv(f"../../output/res_xgb_TargetEncoder_{current_ts}.csv",index=False)
```

# model2 : Logistic Regression

## Bayesian Optimazation to find best hyperparameter for Logistic Regression

```python
pbounds_lr = {"C": [0.1, 10]}

best_param_lr = model.bayesian_opt_logisticRegression(X_train, y_train, pbounds_lr, preprocessor)
```

## fit and predict using Logistic Regression with best hyperparameter.

```python
res_tb_lr = test[["PassengerId"]]
res_tb_lr["Survived"] = model.logisticRegression_fit_and_predict(preprocessor, best_param_lr, X_train, y_train, X_test).astype(int)
res_tb_lr
```

```python
current_ts = model.current_timestamp()
res_tb_lr.to_csv(f"../../output/res_logistic_TargetEncoder{current_ts}.csv",index=False)
```
